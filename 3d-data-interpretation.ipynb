```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random 3-dimensional data
np.random.seed(0)
data = np.random.randn(100, 3)

# Extract x, y, and z coordinates
x = data[:, 0]
y = data[:, 1]
z = data[:, 2]

# Create a scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# Set labels and title
ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')
ax.set_title('3D Scatter Plot')

# Show the plot
plt.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Creating data
np.random.seed(0)
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)

# Creating 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=6,
        color=z,                # set color to an array/list of desired values
        colorscale='Viridis',    # choose a colorscale
        opacity=0.8
    )
)])

# Adding title and labels
fig.update_layout(title='3D Scatter Plot', scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'
))

fig.show()
```

```python
import plotly.graph_objects as go
import numpy as np

# Creating data
np.random.seed(0)
x = np.random.standard_normal(100)
y = np.random.standard_normal(100)
z = np.random.standard_normal(100)
c = np.random.standard_normal(100)  # color of markers
s = np.random.randint(5, 15, 100)  # size of markers

# Creating 3D scatter plot
fig = go.Figure(data=[go.Scatter3d(
    x=x,
    y=y,
    z=z,
    mode='markers',
    marker=dict(
        size=s,  # fourth dimension
        color=c,  # fifth dimension
        colorscale='Viridis',
        opacity=0.8
    )
)])

# Adding title and labels
fig.update_layout(title='5D Scatter Plot', scene=dict(
    xaxis_title='X',
    yaxis_title='Y',
    zaxis_title='Z'
))

fig.show()
```
